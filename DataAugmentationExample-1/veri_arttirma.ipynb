```python
"""**Paketlerin Yüklenmesi**"""

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
img = load_img('logo.jpg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)
```

```python
"""### Kullanılacak veri artırma yönteminin seçilmesi ve hiper parametrelerinin belirlenmesi işlemleri"""

# kullanılacak veri artırma tekniklerini tanımla
datagen = ImageDataGenerator(rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             fill_mode='nearest')
```

`ImageDataGenerator` sınıfını kullanarak görüntü verilerini artırmak için bir veri artırma jeneratorü oluşturur. Bu, mevcut veri setinden yeni ve çeşitli görüntüler oluşturarak yapılır.
- `rotation_range ` Rastgele döndürme açısı aralığı (derece cinsinden). Örneğin, rotation_range=40 ile belirtilirse, görüntüler rastgele olarak -40 ila +40 derece arasında döndürülür.
- `width_shift_range ve height_shift_range` Görüntüyü yatay ve dikey yönde kaydırma aralığı. Örneğin, width_shift_range=0.2 ve height_shift_range=0.2 ile belirtilirse, görüntüler rastgele olarak yatay ve dikey yönde %20'ye kadar kaydırılabilir.
- `shear_range` Kesme(kırpmak) aralığı. Örneğin, shear_range=0.2 ile belirtilirse, görüntüler rastgele olarak %20'ye kadar kesilebilir.
-` zoom_range` Yakınlaştırma aralığı. Örneğin, zoom_range=0.2 ile belirtilirse, görüntüler rastgele olarak %20'ye kadar yakınlaştırılabilir veya uzaklaştırılabilir.
- `horizontal_flip ve vertical_flip` Görüntüyü yatay ve dikey olarak çevirme seçenekleri. Varsayılan olarak False'dir, ancak True olarak ayarlandığında, görüntüler rastgele olarak yatay veya dikey olarak çevrilebilir.

- `fill_mode` Döndürme veya kaydırma sırasında oluşan boş piksellerin doldurulma stratejisi. "nearest" en yakın komşu piksel ile doldurma anlamına gelir.

```python
"""**Artırma işlemi yapılacak olan görüntünün dosyadan okunması işlemi**"""

# tek bir resmi yükle
# logo.jpg dosyasını da deneyebilirsiniz

#img = load_img('ChetBaker.jpg')
img = load_img('logo.jpg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)
```

`img_to_array(img)`
- `img_to_array` fonksiyonu, yüklenen görüntüyü bir NumPy dizisine dönüştürür. Bu, görüntünün piksel değerlerini içeren bir dizi elde etmenizi sağlar. Her bir piksel, görüntünün boyutuna göre bir dizi içinde bulunur. (x, bir dizidir)

`x = x.reshape((1,) + x.shape)` Bu satır, NumPy dizisinin boyutlarını yeniden şekillendirir. Bu örnekte, yüklenen tek bir görüntü olduğu için, dizinin boyutlarına (1,) boyutunu ekler. Bu, daha sonra artırma işlemi için gerekli olan formata dönüştürme işlemini yapar.
- `reshape` Bir NumPy dizisinin boyutlarını ve şeklini yeniden düzenlemek için kullanılan bir yöntemdir.
- `shape` Bir NumPy dizisinin boyutlarını ve şeklini belirten bir özelliktir. Örneğin, bir 2D dizinin şekli (satır sayısı, sütun sayısı) olarak verilir. Bir 3D dizinin şekli ise (yükseklik, genişlik, kanal sayısı) olarak verilir (örneğin, bir renkli görüntünün RGB kanalları için).

```python
import os    # Artirilmis_Veri adında dosya oluşturmak için

if not os.path.exists('Artirilmis_Veri'):
    os.makedirs('Artirilmis_Veri')
```

```python
"""**Giriş görüntüsünden kaç tane üretilmesini istediğimizle ilgili oluşturduğumuz `for` döngüsü ve sonuçların ilgili formatla kaydedilip, dosyaya yazdılırması işlemleri.**"""

# tek resimden 50 tane farklı resim üret ve Artirilmis_Veri klasörüne .jpeg formatında kaydet.
i = 0

for batch in datagen.flow(x, batch_size=1, save_to_dir='Artirilmis_Veri', save_format='jpeg'):
  i+=1

  if i>50:
    break
```

`i = 0` Bir sayaç başlatılır. Bu sayaç, kaç adet artırılmış görüntü üretildiğini takip etmek için kullanılacaktır.

`for batch in datagen.flow(x, batch_size=1, save_to_dir='Artirilmis_Veri', save_format='jpeg')`

- `datagen.flow()` Bu, ImageDataGenerator nesnesinin flow() yöntemini kullanarak artırılmış görüntülerin akışını oluşturur. x giriş görüntüsünü ve diğer parametreleri alır.
- `batch_size=1` Her bir iterasyonda üretilen batch boyutunu belirtir. Burada batch boyutu 1 olarak ayarlanmıştır, yani her bir iterasyonda sadece bir görüntü üretilecektir.
- `save_to_dir='Artirilmis_Veri'` Üretilen artırılmış görüntülerin kaydedileceği dizini belirtir. Bu durumda, görüntüler Artirilmis_Veri adlı bir klasöre kaydedilecektir.
- `save_format='jpeg'` Kaydedilecek görüntülerin formatını belirtir. Bu durumda, görüntüler JPEG formatında kaydedilecektir.
- `i+=1` Her bir iterasyonda, üretilen artırılmış görüntü sayacı bir artırılır.

`if i>50: break` Eğer 50 adet artırılmış görüntü üretildiyse, döngüden çıkılır. Bu şekilde, belirtilen sayıda artırılmış görüntü üretildikten sonra döngü sonlandırılır.
